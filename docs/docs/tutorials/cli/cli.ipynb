with claude code headless mode (claude -p)
https://code.claude.com/docs/en/headless

```python
import dspy
from dspy.clients.cli_lm import CLILM


class SimpleQuery(dspy.Signature):
    question: str = dspy.InputField()
    answer: str = dspy.OutputField()


lm = CLILM("claude -p --dangerously-skip-permissions")
dspy.configure(lm=lm)

predictor = dspy.Predict(SimpleQuery)
question = "What is the biggest file in this directory?"
result = predictor(question=question)

print(f"Question: {question}")
print(f"Claude answer: {result.answer}")
```

and with codex exec

```python
lm = CLILM("codex exec")
dspy.configure(lm=lm)

result = predictor(question=question)

print(f"Question: {question}")
print(f"Codex answer: {result.answer}")
```

```python
lm = CLILM("codex exec --json")
dspy.configure(lm=lm)

result = predictor(question=question)

print(f"Question: {question}")
print(f"Codex answer: {result.answer}")
```
